# 2) How many missing data points do we have?

```python
# total count of missing values in each column
missing_values = sf_permits.isnull().sum()
# lets take a look at the first 10 columns of missing values
missing_values[:10]

# looks like street number suffix is not included in any row
# lets what what percentage of missing cells we have 

# total number of cells 
total_cells = np.product(sf_permits.shape)
# total missing values 
total_missing_values = missing_values.sum()

# percentage of nan or None cells 
nan_or_null_cell_percentage = (total_missing_values/total_cells)*100
nan_or_null_cell_percentage
```
